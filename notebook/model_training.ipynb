```python
import pandas as pd 

# Load the dataset from a CSV file
try:
    df = pd.read_csv('data/Telco-Customer-Churn.csv')
    print("Data loaded successfully!")
    print("First 5 rows of the dataset:")
    print(df.head())
except FileNotFoundError:
    print("Dataset file not found. Please download 'Telco-Customer-Churn.csv' from Kaggle and place it in the correct directory.")
```

## Data Cleaning 

```python
# --- Data Cleaning Steps ---

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# The 'TotalCharges' column might have missing values when tenure is 0.
# It is also an object type and should be numeric.
# Let's inspect rows where TotalCharges might be problematic
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# 'coerce' will turn any non-numeric values into NaN (Not a Number)
# Let's check for NaNs created
print(f"\nNumber of rows with missing TotalCharges after conversion: {df['TotalCharges'].isnull().sum()}")
```

```python
# For customers with 0 tenure, their TotalCharges can be logically filled with 0.
df['TotalCharges'].fillna(0, inplace=True)
print("Missing TotalCharges filled with 0.")
```

```python
# The 'customerID' is not a feature for the model, so we can drop it.
df.drop('customerID', axis=1, inplace=True)
print("\n'customerID' column dropped.")
```

```python
# Our target variable is 'Churn'. Let's convert it to a numerical format (0s and 1s).
df['Churn'] = df['Churn'].apply(lambda x: 1 if x == 'Yes' else 0)
print("'Churn' column converted to numeric (1 for Yes, 0 for No).")
```

```python
print("\nData after cleaning:")
print(df.info())
```
